4.Dotted Line Graph

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('/content/drive/MyDrive/AWP.project.csv','r') as csvfile:
 lines = csv.reader(csvfile, delimiter=',')
 for row in lines:
  x.append(row[0])
  y.append(int(row[1]))

plt.plot(x, y, color = 'g', linestyle = 'dashed',
  marker = 'o',label = "GDP 2020")

plt.xticks(rotation = 0)
plt.xlabel('Countrt')
plt.ylabel('GDP')
plt.title('GDP 2020', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
csvfile.close
```

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open('/content/drive/MyDrive/AWP.project2.csv','r') as csvfile:
 lines = csv.reader(csvfile, delimiter=',')
 for row in lines:
  x.append(row[0])
  y.append(int(row[1]))

plt.plot(x, y, color = 'g', linestyle = 'dashed',
  marker = 'o',label = "GDP 2050 Data")

plt.xticks(rotation = 25)
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP 2050', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```
